# Scapping Dataset

Pada proyek ini, kita akan melakukan analisis sentimen terhadap ulasan dan pendapat pengguna terkait aplikasi "MLBB" di Play Store. Analisis sentimen akan membantu kita untuk memahami bagaimana pengguna merasakan dan mengungkapkan pandangan mereka terhadap aplikasi ini.

Dengan menggunakan berbagai teknik pemrosesan teks dan algoritma machine learning, kita akan mencoba mengidentifikasi apakah ulasan pengguna terhadap aplikasi "MLBB" cenderung positif, negatif, atau netral. Hasil analisis sentimen ini dapat memberikan wawasan berharga kepada pengembang aplikasi untuk meningkatkan pengalaman pengguna. Mari kita mulai menjelajahi ulasan pengguna dan menganalisis sentimen di sekitar aplikasi "MLBB" dengan lebih mendalam.

```python
#Menggali semua ulasan aplikasi dengan ID "" di google Play Store.
scrapeview = reviews_all(
    'com.mobile.legends',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=10000000
)
```

# **Menyimpan ulasan dalam format csv**

## *Dataset saya simpan dalam format .CSV agar bisa di gunakan di notebook yang lain, dataet bisa di dwonload dari menu files di widget google colab*

```python
# Simpan semua kolom dari scrapeview
with open('ulasan_aplikasi_MLBB.csv', mode='w', newline='', encoding='utf-8') as csvfile:
    # Ambil nama kolom dari key dictionary pertama
    fieldnames = scrapeview[0].keys()

    # Gunakan DictWriter agar semua key ditulis sebagai kolom
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # Tulis header kolom
    writer.writeheader()

    # Tulis setiap baris sebagai dictionary
    for row in scrapeview:
        writer.writerow(row)
```
